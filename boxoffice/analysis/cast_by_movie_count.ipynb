```python
from boxoffice.db.frames import get_cast_crew_frame
import matplotlib.pyplot as plt
```

```python
ccf = get_cast_crew_frame()
```

```python
# make a histogram of the number of movies each actor has been in
actor_counts = ccf['person'].value_counts()

# print the histogram values
print(actor_counts)


actor_counts.hist(bins=range(1, 115))

# log scale
plt.yscale('log')

plt.show()
```
