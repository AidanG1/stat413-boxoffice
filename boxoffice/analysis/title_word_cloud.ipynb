```python
from boxoffice.db.frames import get_movie_frame_full
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = get_movie_frame_full()
```

```python
from multiprocessing import process


titles = df['slug']

titles = titles.str.replace('-', ' ')

# remove the apostrophes
titles = titles.str.replace("'", '')
titles = titles.str.replace(".", '')
titles = titles.str.replace(":", '')
titles = titles.str.replace(";", '')
titles = titles.str.replace(",", '')
titles = titles.str.replace("!", '')


sep = ' '

titles = titles.str.cat(sep=sep)

wc = WordCloud(
    background_color='white',
    max_words=150,
    width=800,
    height=400,
    colormap='tab20b',
).generate_from_text(titles)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()

# print the processed titles
processed = WordCloud().process_text(titles)

# sort the processed titles
sorted_titles = sorted(processed.items(), key=lambda x: x[1], reverse=True)

# print the top 10
print(sorted_titles[:10])
```
